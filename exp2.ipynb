```python
'''
Aim:
To apply preprocessing techniques to clean, transform, and organize raw data into a structured format suitable for accurate and efficient analysis or model building.

Description:
Preprocessing techniques are essential steps in preparing raw data for analysis, modeling, or machine learning tasks. In real-world scenarios, data is often incomplete, inconsistent, and noisy, which can negatively impact the performance of algorithms. Therefore, preprocessing focuses on improving data quality and transforming it into a structured and usable format.

The process typically begins with data cleaning, where missing values are handled by methods such as imputation or removal, and inconsistencies or errors are corrected. Noise and outliers are also identified and treated to prevent misleading results. Another important step is data transformation, which includes normalization or scaling to bring all features into a similar range, making computations more efficient and accurate.

Additionally, preprocessing may involve encoding categorical variables into numerical form, aggregating data, and reducing dimensionality to simplify complex datasets. These steps help in reducing redundancy and improving the overall efficiency of data processing. By applying preprocessing techniques, the dataset becomes more reliable, consistent, and suitable for further analysis, ultimately leading to better performance and more accurate outcomes in analytical or machine learning applications.

'''
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer

# Load dataset (change file name as needed)
df = pd.read_csv("raw_merged_heart_dataset.csv")

# Display first 5 rows
print("Original Data:")
print(df.head())

# -----------------------------
# 🔹 1. DATA CLEANING
# -----------------------------

# Remove duplicate rows
df = df.drop_duplicates()

# Handle missing values

# For numerical columns → fill with mean
num_imputer = SimpleImputer(strategy='mean')
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = num_imputer.fit_transform(df[num_cols])

# For categorical columns → fill with most frequent value
cat_imputer = SimpleImputer(strategy='most_frequent')
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

# -----------------------------
# 🔹 2. FEATURE ENCODING
# -----------------------------

# Label Encoding (for ordinal or simple categories)
label_encoder = LabelEncoder()

for col in cat_cols:
    df[col] = label_encoder.fit_transform(df[col])

# OR use One-Hot Encoding (better for non-ordinal data)
# df = pd.get_dummies(df, columns=cat_cols)

# -----------------------------
# 🔹 FINAL OUTPUT
# -----------------------------

print("\nCleaned & Encoded Data:")
print(df.head())

# Save cleaned dataset
df.to_csv("cleaned_data.csv", index=False)
```
